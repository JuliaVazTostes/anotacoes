# **CONSULTA UTILIZANDO O QUERY:**

O Query permite utilizar operadores lógicos, como:

AND (&);
OR (|);
FALSE (!);

Com este auxílio é possível formar as mais diversificadas queries. É possível também verificar elementos dentro de um array com o IN, atingindo assim resultados mais específicos.

```
# Consulta mais básica que se pode fazer utilizando o query: UMA CONDIÇÃO -> Solicitando os dados em que o AveragePrice é menor ou igual a 1

df.query('AveragePrice <= 1')
```

```
df.Total_Volume
```

```
# Com espaço

df['Total_Volume'] 
```

```
# Utilizar operador diferente (!=) Verificar 'type' diferente de 'conventional'

df.query('type != "conventional"')
```

```
# DUAS CONDIÇÕES ->Podemos adicionar o operador &, o que possibilita adicionar mais uma condição, tornando a consulta mais específica.
# Verificar dados com AveragePrice maior que 1 e Total_Volume < 100000

df.query('AveragePrice > 1 & Total_Volume < 100000')
```

```
df.columns
```

```
# Utilizando diferente (!=) e igual (==) -> Selecionar dados que não são da Albania e são de 2016

df.query('region != "Albany" & year == 2016')
```

```
# Utilizando diferente OU (|) -> Verificando Average Price maior que 1.50 ou Large_Bags maior que 200000

df.query('AveragePrice == 1 | Large_Bags > 200000')
```

```
# UTILIZAÇÃO DO IN: Consultar por resultados especifícos -> Consultar dados apenas dos anos 2015 e 2017

df.query('year in [2015,2017]')
```
